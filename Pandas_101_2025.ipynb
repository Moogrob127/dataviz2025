# HW 4 (group)

ยอดรวมทั้ง 5 เดือน เมืองไหนมีจำนวนผู้ถูกเรียกค้นที่เป็นคน Asia น้อยที่สุด และ มากที่สุด



````
import pandas as pd
import glob
import os

# 1. ทำการเชื่อมต่อกับ Google Drive (เหมือนเดิม)
from google.colab import drive
drive.mount('/content/drive')

# 2. กำหนด Path หลักและรูปแบบของไฟล์ที่ต้องการค้นหา
# เราจะใช้เครื่องหมาย * (Wildcard) แทนส่วนของชื่อไฟล์หรือโฟลเดอร์ที่เปลี่ยนแปลงไป
# รูปแบบ: ค้นหาไฟล์ .csv ที่อยู่ในโฟลเดอร์ 2025-XX ใดๆ ก็ตาม
path_pattern = '/content/drive/MyDrive/Data/2025-*/*-stop-and-search.csv'
all_stopandsearch = glob.glob(path_pattern)

print(f"ค้นพบไฟล์ CSV ทั้งหมด: {len(all_stopandsearch)} ไฟล์")
# ลองแสดงตัวอย่าง 5 ไฟล์แรกที่เจอ
print("ตัวอย่างไฟล์ที่พบ:")
for f in all_stopandsearch[:5]:
    print(f)
````

````
# 3. สร้าง Loop เพื่ออ่านไฟล์ทั้งหมดและรวมเป็น DataFrame เดียว
all_dfs = [] # สร้าง list ว่างเพื่อเก็บ DataFrame ชั่วคราว

for filepath in all_stopandsearch:
    # อ่านไฟล์ CSV จาก path ที่ได้มา
    df = pd.read_csv(filepath)

    # --- ส่วนสำคัญ: ดึงชื่อเขตตำรวจจากชื่อไฟล์ ---
    # ชื่อไฟล์: '.../2025-01-avon-and-somerset-stop-and-search.csv'
    # 1. ดึงเฉพาะชื่อไฟล์ออกมา: os.path.basename(filepath) -> '2025-01-avon-and-somerset-stop-and-search.csv'
    filename = os.path.basename(filepath)
    # 2. แยกส่วนประกอบของชื่อไฟล์ด้วยเครื่องหมาย '-'
    parts = filename.split('-')
    # 3. ชื่อเขตตำรวจจะอยู่ระหว่างส่วนของวันที่ (2 ส่วนแรก) และ 3 ส่วนท้าย ('stop', 'and', 'search.csv')
    force_name_parts = parts[2:-3]
    # 4. นำชิ้นส่วนชื่อมารวมกันและจัดรูปแบบตัวพิมพ์ให้สวยงาม (เช่น 'avon and somerset' -> 'Avon And Somerset')
    force_name = ' '.join(force_name_parts).title()

    # เพิ่มคอลัมน์ 'Police Force' เข้าไปใน DataFrame
    df['Police Force'] = force_name

    all_dfs.append(df)

# รวม DataFrame ทั้งหมดใน list ให้เป็น DataFrame เดียว
combined_df = pd.concat(all_dfs, ignore_index=True)

print("รวมข้อมูลจากทุกไฟล์สำเร็จแล้ว!")
print(f"DataFrame ใหม่มีข้อมูลทั้งหมด: {len(combined_df):,} แถว")
print(f"จากเขตตำรวจทั้งหมด: {combined_df['Police Force'].nunique()} เขต")
````

````
# --- จากตรงนี้ สามารถนำโค้ดวิเคราะห์จากคำตอบที่แล้วมาใช้ต่อได้เลย ---

# คัดกรองข้อมูลเฉพาะผู้ถูกเรียกค้นที่เป็นคน Asia
asian_stops_df = combined_df[combined_df['Self-defined ethnicity'].str.contains('Asian', na=False)]

# จัดกลุ่มและนับจำนวนตามเขตตำรวจ
asian_counts = asian_stops_df['Police Force'].value_counts()

print("สรุปผลการเรียกค้นบุคคลเชื้อสายเอเชียในรอบ 5 เดือน (จากข้อมูลที่โหลดแบบอัตโนมัติ):\n")
print(f"📈 เขตที่มีการเรียกค้นมากที่สุด: {asian_counts.idxmax()} ({asian_counts.max():,} ครั้ง)")
print(f"📉 เขตที่มีการเรียกค้นน้อยที่สุด: {asian_counts.idxmin()} ({asian_counts.min():,} ครั้ง)")
````

# Hw5

จาก Hw4 ให้แก้จากโค้ดเดิม เพื่อหายอดรวมทุกเมือง เดือนไหนที่คน Asian ถูกหยุดเรียกค้นมากที่สุด และน้อยที่สุด

````
import pandas as pd
import glob
import os

# 1. ทำการเชื่อมต่อกับ Google Drive (เหมือนเดิม)
from google.colab import drive
drive.mount('/content/drive')

# 2. กำหนด Path หลักและรูปแบบของไฟล์ที่ต้องการค้นหา
# เราจะใช้เครื่องหมาย * (Wildcard) แทนส่วนของชื่อไฟล์หรือโฟลเดอร์ที่เปลี่ยนแปลงไป
# รูปแบบ: ค้นหาไฟล์ .csv ที่อยู่ในโฟลเดอร์ 2025-XX ใดๆ ก็ตาม
path_pattern = '/content/drive/MyDrive/Data/2025-*/*-stop-and-search.csv'
all_stopandsearch = glob.glob(path_pattern)

print(f"ค้นพบไฟล์ CSV ทั้งหมด: {len(all_stopandsearch)} ไฟล์")
# ลองแสดงตัวอย่าง 5 ไฟล์แรกที่เจอ
print("ตัวอย่างไฟล์ที่พบ:")
for f in all_stopandsearch[:5]:
    print(f)
````

````
dfs = [] # สร้าง list ว่างเพื่อเก็บ DataFrame ชั่วคราว

for filepath in all_stopandsearch:
    # อ่านไฟล์ CSV จาก path ที่ได้มา
    df = pd.read_csv(filepath)

    # ดึงชื่อไฟล์ เช่น '2025-01-avon-and-somerset-stop-and-search.csv'
    filename = os.path.basename(filepath)
    # แยกเดือนจากชื่อไฟล์
    month = '-'.join(filename.split('-')[:2])
    # เพิ่มคอลัมน์เดือนเข้าไปใน DataFrame
    df['month'] = month

    # เก็บเฉพาะแถวที่มี 'Asian'
    if 'Self-defined ethnicity' in df.columns:
        df_asian = df[df['Self-defined ethnicity'].astype(str).str.contains('Asian', case=False, na=False)]
        dfs.append(df_asian)

# รวมทุก DataFrame เข้าเป็นตัวเดียว
combined_df = pd.concat(dfs, ignore_index=True)

# นับจำนวนตามเดือน
month_counts = combined_df['month'].value_counts()

# แสดงผล
print("สรุปผลการเรียกค้นบุคคลเชื้อสายเอเชียในแต่ละเดือน (จากข้อมูลที่โหลดแบบอัตโนมัติ):\n")
print(f"\n📈 เดือนที่ค้นตัวคน Asian มากที่สุด: {month_counts.idxmax()} ({month_counts.max():,} ครั้ง)")
print(f"📉 เดือนที่ค้นตัวคน Asian น้อยที่สุด: {month_counts.idxmin()} ({month_counts.min():,} ครั้ง)")
````

````
# prompt: คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่อง ว่า แม่ หรือ พ่อ มากที่สุด (พร้อมทั้งอธิบาย code แต่ละบรรทัดด้วย)

# สร้าง dictionary เพื่อเก็บจำนวนนิทานที่มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่องสำหรับแต่ละคนแต่ง
creator_counts = {}
for index, row in this_data.iterrows():
  if 'แม่' in row['title'] or 'พ่อ' in row['title']:
    creator = row['creator']
    if creator not in creator_counts:
      creator_counts[creator] = 0
    creator_counts[creator] += 1

# หาคนแต่งที่มีจำนวนนิทานมากที่สุด
max_count = 0
most_prolific_creator = None
for creator, count in creator_counts.items():
  if count > max_count:
    max_count = count
    most_prolific_creator = creator

print(f"The creator who wrote the most stories with 'แม่' or 'พ่อ' in the title is {most_prolific_creator} with {max_count} stories.")
````
