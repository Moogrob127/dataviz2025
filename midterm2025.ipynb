เธอไปคอนเสิร์ต SQL Symphony 3 ครั้งในเดือนธันวาคม 2017

```python
facebook_checkins_df = pd.read_csv('/content/drive/MyDrive/Exam_Data_DataViz2025/facebook_event_checkin.csv')
display(facebook_checkins_df.head())
```

```python
# กรองข้อมูล event_name = SQL Symphony Concert ในเดือนธันวาคม 2017
sql_symphony_checkins_december = facebook_checkins_df[
    (facebook_checkins_df["event_name"] == "SQL Symphony Concert") &
    (facebook_checkins_df["date"].astype(str).str.startswith('201712'))
]

# นับดูว่า person_id ไหนบ้างที่ซ้ำกัน 3 รอบ
sql_symphony_counts = sql_symphony_checkins_december['person_id'].value_counts()
mastermind_person_id = sql_symphony_counts[sql_symphony_counts == 3].index

# ดูว่า person_id ที่ซ้ำ 3 รอบ คนนั้นเป็นใคร โดยเทียบกับ id ในตาราง person_df
mastermind = person_df[person_df['id'].isin(mastermind_person_id)]
display(mastermind)
```
