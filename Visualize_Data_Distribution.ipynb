# การแสดงผลการกระจายของข้อมูล

````
import pandas as pd
from matplotlib import pyplot as plt
````

````
from google.colab import drive
drive.mount('/content/drive')
````

````
stopandsearch_df = pd.read_csv('/content/drive/MyDrive/Data/2025-01/2025-01-bedfordshire-stop-and-search.csv')
stopandsearch_df.head()
````

แปลงค่า Gender และ Age range เป็นจำนวนจริง

````
stopandsearch_df.groupby('Gender').count()
````

````
stopandsearch_df.groupby('Age range').count()
````

````
stopandsearch_df.shape
````

````
stopandsearch_df_dropna = stopandsearch_df.dropna(subset=['Age range','Gender'])
stopandsearch_df_dropna.head()
````

````
stopandsearch_df_dropna['Gender'] = stopandsearch_df_dropna['Gender'].replace({'Male':35,'Female':1})
stopandsearch_df_dropna.head()
````

````
stopandsearch_df_dropna['Age range'] = stopandsearch_df_dropna['Age range'].replace({'under 10': 10 , 'over 34': 34 , '25-34':30 , '18-24':21 , '10-17':14 })
stopandsearch_df_dropna.head()
````

## สร้าง list ของ  random number ที่เป็น normal distribution

### Histogram

````
import numpy as np
list_of_rn = np.random.normal(loc=1, scale=0.1, size=len(stopandsearch_df_dropna))
````

````
plt.hist(list_of_rn)
````

````
plt.hist(list_of_rn)
````

````
hist_out = plt.hist(list_of_rn)
````

````
hist_out
````

````
hist_out[0]
````

````
hist_out = plt.hist(list_of_rn,bins=20)
````

````
stopandsearch_df_dropna['Gender'] = stopandsearch_df_dropna['Gender'] * list_of_rn
````

````
stopandsearch_df_dropna['Age range'] = stopandsearch_df_dropna['Age range'] * np.random.normal(loc=1, scale=0.01, size=len(stopandsearch_df_dropna))
````

# Quiz11

plot กราฟ 3 อันในแนวทแยงมุม
1. boxploy
2. scatter
3. histogram

````
fig, axes = plt.subplots(3, 3, figsize=(15, 15))

# Boxplot
axes[0, 0].boxplot(stopandsearch_df_dropna['Age range'].dropna())
axes[0, 0].set_title('Boxplot of Age Range')

# Scatter plot
axes[1, 1].scatter(stopandsearch_df_dropna['Age range'], stopandsearch_df_dropna['Gender'], alpha=0.5)
axes[1, 1].set_title('Scatter Plot of Age Range vs Gender')
axes[1, 1].set_xlabel('Age Range')
axes[1, 1].set_ylabel('Gender')

# Histogram
axes[2, 2].hist(stopandsearch_df_dropna['Age range'].dropna())
axes[2, 2].set_title('Histogram of Age Range')

plt.tight_layout()
plt.show()
````
